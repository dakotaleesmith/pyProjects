```python
### IMPORTING MODULES
from decouple import config
import pandas as pd
import duolingo
import json
```

```python
### DECLARING CREDENTIALS AND CONNECTING TO API
username = config('DuoUser', default='defaulted')
password = config('DuoPass', default='defaulted')
lingo = duolingo.Duolingo(username, password)
```

```python
abbr = lingo.get_abbreviation_of('Spanish')

userInfo = lingo.get_user_info()
df = pd.json_normalize(userInfo)
```

```python
### WRITING COLUMN NAMES TO A TEXT FILE FOR READABILITY'S SAKE
with open("user_info_columns.txt", "w") as f:
    for column in df.columns:
        f.write(f"{column}\n")
```
